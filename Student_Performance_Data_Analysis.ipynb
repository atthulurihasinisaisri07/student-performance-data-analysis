```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df.value_counts()
df.groupby('gender')['math score'].mean()
```

```python
sns.barplot(x='gender', y='math score', data=df)
plt.show()

```

```python
sns.histplot(x='gender', y='math score', data=df)
plt.show()
```

```python
df[['math score', 'reading score', 'writing score']].mean().plot(kind='bar')
plt.title("Average Scores of Students")
plt.ylabel("Average Marks")
plt.xlabel("Subjects")
plt.show()
```

```python
sns.barplot(x='gender', y='math score', data=df)
plt.title("Gender vs Math Score")
plt.show()
```

```python
sns.histplot(df['math score'], kde=True)
plt.title("Math Score Distribution")
plt.show()
```

```python
df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean()
```

```python
df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean().plot(kind='bar')
plt.title("Impact of Test Preparation Course on Scores")
plt.ylabel("Average Score")
plt.xlabel("Test Preparation Course")
plt.show()
```

# Conclustion
In this project, student performance data was analyzed using Python. The dataset contained 1000 student records with academic scores and background information. Data exploration and visualization revealed that students performed best in reading, while math was the most challenging subject. Gender-wise differences in performance were minimal. Students who completed the test preparation course scored significantly higher across all subjects, indicating the positive impact of structured preparation on academic outcomes.
